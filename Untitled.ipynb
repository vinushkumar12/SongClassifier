```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold, cross_val_score
```

```python
songs = pd.read_csv("/Users/vinushkumar/Desktop/Audio Classifier/fma-rock-vs-hiphop.csv")
print(songs)
```

```python
songs.head()
```

```python
metrics = pd.read_json("/Users/vinushkumar/Desktop/Audio Classifier/echonest-metrics.json")
```

```python
metrics.head()
```

```python
merged_songs = pd.merge(left = songs[['track_id', 'genre_top']], right=metrics, on='track_id')
merged_songs.head()
```

```python
merged_songs.head()
```

```python
correlations = merged_songs.corr()
print(correlations)
```

```python
correlations.style.background_gradient()
```

```python
features = merged_songs.drop(['track_id', 'genre_top'], axis = 1)
labels = merged_songs['genre_top']
print(features)
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(features)
```

```python
scaled
```

```python
pca = PCA()
pca.fit(scaled)
explained = pca.explained_variance_ratio_
explained
```

```python
fig, ax = plt.subplots()
ax.bar(range(pca.n_components_), explained)
ax.set_xlabel("Principal Componenet #")
ax.set_ylabel("Explained Variance")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(range(8), np.cumsum(explained))
ax.axhline(y=0.9, linestyle='--')
plt.show
```

```python
n = 7
pca = PCA(n, random_state=10)
pca.fit(scaled)
pca_prediction = pca.transform(scaled)
pca_prediction
pca_prediction.shape
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(pca_prediction, labels, stratify = labels)
tree = DecisionTreeClassifier(random_state = 10)
tree.fit(train_features, train_labels)
tree_rules = export_text(tree, feature_names=[f'Feature {i}' for i in range(train_features.shape[1])])
print(tree_rules)
```

```python
predictive_tree = tree.predict(test_features)
accuracy_score(test_labels, predictive_tree)
```

```python
labels.value_counts
```

```python
confusion_matrix(test_labels, predictive_tree)
```

```python
logis = LogisticRegression(random_state = 10)
logis.fit(train_features, train_labels)
pred_label_logs = logis.predict(test_features)
accuracy_score(test_labels, pred_label_logs)
```

```python
trees_class = classification_report(test_labels, predictive_tree)
logis_class = classification_report(test_labels, pred_label_logs)
print(trees_class)
print(logis_class)
```

```python
hip = merged_songs.loc[merged_songs['genre_top'] == 'Hip-Hop']
rock = merged_songs.loc[merged_songs['genre_top'] == 'Rock']
rock.head()
```

```python
rock = rock.sample(n = hip.shape[0])
rock.shape
```

```python
hip.shape
```

```python
new_data = pd.concat([rock,hip])
```

```python
print(new_data)
```

```python
features = new_data.drop(['track_id', 'genre_top'], axis = 1)
labels = new_data['genre_top']
scalar = StandardScaler()
scalar = scaler.fit_transform(features)
pca = PCA()
pca.fit(scalar)
explained = pca.explained_variance_ratio_
fig, ax = plt.subplots()
ax.bar(range(pca.n_components_), explained)
ax.set_xlabel("Principal Componenet #")
ax.set_ylabel("Explained Variance")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(range(8), np.cumsum(explained))
ax.axhline(y=0.9, linestyle='--')
plt.show
```

```python
n = 7
pca = PCA(n, random_state = 10)
pca.fit(scalar)
pca_prediction = pca.transform(scalar)
pca_prediction.shape
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(pca_prediction, labels, stratify = labels)
tree = DecisionTreeClassifier(random_state = 10)
tree.fit(train_features, train_labels)
pred = tree.predict(test_features)
accuracy_score(test_labels, pred)
```

```python
logistic = LogisticRegression(random_state = 10)
logistic.fit(train_features, train_labels)
pred_log = logistic.predict(test_features)
accuracy_score(test_labels, pred_log)
```

```python
trees_class = classification_report(test_labels, pred)
logis_class = classification_report(test_labels, pred_log)
print(trees_class)
print(logis_class)
```

```python
kf = KFold(n_splits=10)
tree = DecisionTreeClassifier(random_state=10)
logs = LogisticRegression(random_state=10)
tree_scores = cross_val_score(tree, pca_prediction, labels, cv=kf)
log_score = cross_val_score(logs, pca_prediction, labels, cv = kf)
print("Decision Tree:", np.mean(tree_scores))
print("Log_score:", np.mean(log_score))
```
